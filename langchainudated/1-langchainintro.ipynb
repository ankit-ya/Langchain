# Langchain Version v1

```python
import os
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv())

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.agents import create_agent


def get_weather(city: str) -> str:
    """Get the weather of a city."""
    return f"The weather of {city} is 30 degree celsius"


# NOTE: this is at column 0, so it runs at the top level (NOT inside the function)
agent = create_agent(
    model="groq:llama-3.3-70b-versatile",
    tools=[get_weather],
    system_prompt=(
        "You are a helpful weather assistant. "
        "Whenever the user asks about the weather in a city, you MUST call the "
        "get_weather tool and report exactly what it returns. "
        "Do not add disclaimers or tell the user to check other websites."
    ),
)
```

```python
# run the agent  (key is "messages", plural)
response = agent.invoke(
    {"messages": [{"role": "user", "content": "What is the weather of New York? AND HAVE ANY CHANCE OF RAIN? ALSO TELLME THEIR CURRENT TEMPERATURE."}]}
)

# the last message is the agent's reply; .content is an attribute
print(response["messages"][-1].content)
```

Model Integreation with OpenAI , Google Geminin and GROQ 

```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
from langchain.chat_models import init_chat_model
model = init_chat_model( model="groq:llama-3.3-70b-versatile")

response = model.invoke("Hello how are you?")
print(response.content)
```

Streaming AND Batch :-

Streaming :- Most Model can stream their Output contents while its is being generated. By Displaying output Progresssively , streaming significantly improve user exprience , particulary for longer repsonse ,calling stream() returns an iterator that yield output chunks as they are produces . you can use a loop to process each chunk in real time .

```python
model.invoke("What is the weather of New York? AND HAVE ANY CHANCE OF RAIN? ALSO TELLME THEIR CURRENT TEMPERATURE.")
```

```python
for chunk in model.stream("What is the weather of New York? AND HAVE ANY CHANCE OF RAIN? ALSO TELLME THEIR CURRENT TEMPERATURE."):
    print(chunk.content, end="")
```

Batch :- Batching a collection of independents request to a moel can significatly improve performance and reduce costs , as the processing can be done in parallel .

```python
response = model.batch([
    "What is the weather of New York?",
     "How do airplan fly?",
      "What is quantum computing?" ])
for response in response:
    print(response.content)
```
